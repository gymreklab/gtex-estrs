```python
tissue = TISSUE[0]
data = pd.read_csv(MASTR, sep='\t')
data['anova.percentile'] = pd.qcut(data['qvalue'].rank(method='first').values, 5, labels=False)
data['gcta.percentile'] = pd.qcut(data['cis_str_h2'], 5, labels=False)
data['caviar.percentile'] = pd.qcut(data['best.str.score'], 5, labels=False)
data['causal.percentile'] = pd.qcut(data['top.variant.score'], 5, labels=False)

VIEW = ['gene', 'chrom', 'gene.name', 'str.id', 'qvalue','cis_str_h2','best.str.score','causality.score','anova.percentile','gcta.percentile','caviar.percentile','causal.percentile','top.variant.score','logL']
Percentiles = data[VIEW]
```

```python
for j in range(5):
    decile1 = Percentiles.loc[Percentiles['anova.percentile']==j]
    try:
        print('\n In the ',j+1,'th decile of ANOVA test where max and min qvalues are respectively ',max(decile1['qvalue']),' and ', min(decile1['qvalue'])) 
    except:
        print('No data in ',j+1,'th quintile')
        continue
    for i in range(10):
        count=max(decile1.loc[decile1['gcta.percentile']==i]['cis_str_h2'])
        print(decile1.loc[decile1['gcta.percentile']==i].shape[0], ' are in the ', i+1,'th decile in gcta analysis    ie       cis_str_h2<=',count)
    
```
